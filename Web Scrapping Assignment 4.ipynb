# 5
Scrape details of Trending repositories on Github.com
You have to find the following details:
    
    a. Reporsitory Title
    b. Repository Description
    c. Contributor count
    d. Language Used

```python
driver = webdriver.Chrome('C:\Python\Chrome\chromedriver.exe')
driver.set_window_size(1500,1000)
```

```python
driver.get('https://github.com/')
print(driver.title)
try:
    optn = driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]/a')
    driver.get(optn.get_attribute('href'))
    print(driver.title)
    
except NoSuchElementException:
    Print("Web Element not Found")
```

```python
df=pd.DataFrame()
column=["Repository Title", "Repository Descr","Contributor Count","Language Used"]

for a in driver.find_elements_by_xpath('/html/body/div[4]/main/div[3]/div/div[2]/article'):
    Title=a.find_element_by_xpath('.//h1/a').text
    Descr=a.find_element_by_xpath('.//p').text
    Count=len(a.find_elements_by_xpath('.//div[2]/span[2]/a'))
    try:
        Lang=a.find_element_by_xpath('.//div[2]/span[1]/span[2]').text
    except NoSuchElementException:
        Lang=""
    rows=[Title,Descr,Count, Lang]
    df=df.append(pd.DataFrame([rows],columns=column))
    
df
```

    e. Top Position
    f. Weeks on chart

```python
driver = webdriver.Chrome('C:\Python\Chrome\chromedriver.exe')
driver.set_window_size(1200,1000)
```

```python
driver.get('https://www.billboard.com/')
print(driver.title)
try:
    optn = driver.find_element_by_xpath('/html/body/div[3]/header/div[2]/div/nav/ul/li[1]/a')
    driver.get(optn.get_attribute('href'))
    print(driver.title)
    
except NoSuchElementException:
    Print("Web Element not Found")
```

```python
df=pd.DataFrame()
column=["Rank", "Song","Artist","Last Week Position", "Peak Position","Weeks on Chart"]
for a in driver.find_elements_by_xpath('/html/body/div[4]/main/div[2]/div[3]/div/div/div/div[2]/div[@class="o-chart-results-list-row-container"]'):
    rank=a.find_element_by_xpath('.//ul/li[1]/span[1]').text
    song=a.find_element_by_xpath('.//ul/li[4]/ul/li[1]/h3').text
    artist=a.find_element_by_xpath('.//ul/li[4]/ul/li[1]/span').text
    lastweek=a.find_element_by_xpath('.//ul/li[4]/ul/li[4]/span').text
    toppos=a.find_element_by_xpath('.//ul/li[4]/ul/li[5]/span').text    
    weeks=a.find_element_by_xpath('.//ul/li[4]/ul/li[6]/span').text 
    rows=[rank, song, artist, lastweek, toppos, weeks]
    df=df.append(pd.DataFrame([rows],columns=column))
```

Note: From naukri.com homepage click on recruiters option and on the search pane type Data Science and click on search. This should all be done through code.


```python
driver = webdriver.Chrome('C:\Python\Chrome\chromedriver.exe')
driver.set_window_size(1200,1000)
```

```python
driver.get('https://www.naukri.com/hr-recruiters-consultants')
print(driver.title)
search=driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[1]/div[1]/form/div[1]/div/div[1]/div[1]/div[2]/input')
search.send_keys("Data Science")
search=driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[1]/div[1]/form/div[1]/button')
search.click()
```

```python
df=pd.DataFrame()
column=["Recruiter Name", "Recruiter Designation","Company","Skills they hire","Recruiter Location"]

data=driver.find_element_by_xpath('/html/body/div[3]/div/div[2]/div/div[@id="tabP-1"]')
for r in data.find_elements_by_xpath('.//div[@class="recInfo"]'):
    info = r.find_elements_by_xpath('.//span')
    name= info[0].text
    desig = info[1].text
    try:
        company = r.find_elements_by_xpath('.//div/p/a[@class="ellipsis"]')[1].text
    except (NoSuchElementException, IndexError):
        company=""
        
    try:
        location = info[2].text
    except IndexError:
        location=""
    skill = r.find_element_by_xpath('.//div[@class="hireSec highlightable"]').text
    rows=[name, desig, company, skill, location]
    df=df.append(pd.DataFrame([rows],columns=column))
```

## 9. 
Scrape the details most watched tv series of all time from imdb.com. <br> Url = https://www.imdb.com/list/ls095964455/ 

You have to find the following details:

|Name|Year span|Genre|Run time|Ratings|Votes|
|---|---|---|---|---|---|

```python
df=pd.DataFrame()
column= ["Name","Year Span", "Genre", "Runtime", "Rating","Votes"]
r=requests.get('https://www.imdb.com/list/ls095964455/', verify=False)
soup=BeautifulSoup(r.text,'html.parser')
soup=soup.find_all("div", {"class" : "lister-item-content"})
for item in soup:
    Name = item.h3.a.text
    YearSpan = item.h3.find_all('span')[1].text
    Runtime=item.p.find('span',{"class":"runtime"}).text
    Genre=item.p.find('span',{"class":"genre"}).text
    Rating=item.div.div.span.text
    Votes=item.find('span',{"name":"nv"}).text
    rows = [Name.strip(), YearSpan.strip(), Genre.strip(), Runtime.strip(), Rating.strip(), Votes.strip()]
    df=df.append(pd.DataFrame([rows], columns=column))
HTML(df.style.to_html())    
```
